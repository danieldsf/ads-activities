```python
import pandas as pd
df = pd.read_csv('train.csv')

```

```python
# A
sobreviventes = df[df['Survived'] == 1]
print(sobreviventes)
```

```python
# B
def ver_percentual(todo, parte):
    return '%.2f%%' % ((len(parte)/len(todo))*100)
print(ver_percentual(df, sobreviventes))
```

```python
# C
for i in df['Pclass'].unique():
    sobreviventes_mulheres = df.query("Sex == 'female' and Pclass == '%s' and Survived == '1' "%i)
    mulheres = df.query("Sex == 'female' and Pclass == '%s'"%i)
    print(ver_percentual(mulheres, sobreviventes_mulheres))
```

```python
# D
#deleted  = df[df['Embarked'].isnull()]

filtered_df = df[df['Embarked'].notnull()]
#df.drop(deleted.index, axis=1)
#deleted
deleted  = filtered_df[filtered_df['Embarked'].isnull()]
#deleted
#for i in df[df['Embarked'].isnull()]:
#    print(i)
```

```python
# E

def alterSex(sex):
    sex = sex.replace("female", "0").replace("male", "1")
    return sex
df['Sex'] = df['Sex'].apply(alterSex)
df
```

```python
# F
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
#df
df['IsAlone'] = df['FamilySize'] > 1

def alterAlone(a):
    if(a):
        return 1
    return 0
df['IsAlone'] = df['IsAlone'].apply(alterAlone)

df
```

```python
# G
x_cat = pd.get_dummies(df[['Embarked']])
x_cat
```

```python
# H
df['Embarked'] = x_cat
x = df[['Sex', 'IsAlone','Embarked', 'Pclass']]
y = df['Survived']
```

```python
# I
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.20,random_state=0)

modelo = MultinomialNB()

modelo.fit(x_treino, y_treino)

modelo.score(x_teste, y_teste)
```

```python
# J.1
x_sem = x
x_sem = x_sem.drop(columns="Pclass")
x_sem

# De novo

x_treino, x_teste, y_treino, y_teste = train_test_split(x_sem, y, test_size=0.20,random_state=0)

modelo = MultinomialNB()

modelo.fit(x_treino, y_treino)

modelo.score(x_teste, y_teste)
```

```python
# J.1
def alterClass(c):
    lista = ['A', 'B', 'C']
    #sex = sex.replace("female", "0").replace("male", "1")
    return lista[int(c)-1]


x_cat2 = x
x_cat2['Pclass'] = x_cat2['Pclass'].apply(alterClass)
x_cat2['Pclass'] = pd.get_dummies(x_cat2[['Pclass']])
#x_cat2
# De novo

x_treino, x_teste, y_treino, y_teste = train_test_split(x_cat2, y, test_size=0.20,random_state=0)

modelo = MultinomialNB()

modelo.fit(x_treino, y_treino)

modelo.score(x_teste, y_teste)
```
